# Crosstab

```python
import pandas as pd
df=pd.read_csv("crosstab.csv")
df
```

```python
pd.crosstab(df.Nationality,df.Handness) # first argument is row (df.Nationality) second argument is column value
```

```python
pd.crosstab(df.Age,df.Handness,margins=True)
```

```python
pd.crosstab([df.Nationality,df.Age],df.Handness,margins=True) #first list is for row
```

```python
pd.crosstab(df.Nationality,[df.Age,df.Handness],margins=True) # multiple   column can be list
```

```python
pd.crosstab(df.Nationality,[df.Handness],normalize=True) #printing percentage of each value
```

```python
import numpy as np
pd.crosstab(df.Nationality, df.Handness, values=df.Age, aggfunc=np.average) # printing average age for right and left hand
```
